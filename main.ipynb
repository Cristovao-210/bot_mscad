```python
## ABRINDO NAVEGADOR E CONECTANDO NO ESIAPE ##
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

service = Service(ChromeDriverManager().install())
nav = webdriver.Chrome(service=service)
nav.get("https://esiape.sigepe.gov.br/")
```

```python
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from selenium.webdriver import ActionChains
import pandas as pd
#import pyperclip

# Estrutura de dados para armazenar matrículas
dicionario_siapeCAD = {}

# DIGITANDO A MATRÍCULA SIAPE PARA ACESSAR CADASTRO
servidores = pd.read_csv('servidores606_2.csv', sep=',', encoding='latin-1')
for matricula in servidores['SIAPE']:
    
    try:
        
        nav.switch_to.default_content()
        iframe = nav.find_element(By.NAME,'WA0') 
        nav.switch_to.frame(iframe)
        time.sleep(1)
        nav.find_element(By.ID, 'F_131').click()
        # # escrever a matrícula
        nav.find_element(By.XPATH, '//*[@id="F_131"]').send_keys(str(matricula).strip())
        # # precionar enter depois de colocar a matrícula
        time.sleep(1)
        ActionChains(nav).send_keys(Keys.ENTER).perform()
        # # pausa para o sistema 'pensar'
        time.sleep(3)
        # # PRESCIONANDO 'ENTER' PARA AVANÇAR A PÁGINA
        ActionChains(nav).send_keys(Keys.SPACE).perform()
        time.sleep(1)
        ActionChains(nav).send_keys(Keys.ENTER).perform()
        time.sleep(3)
        
        # COPIANDO NOME E MATRÍCULA SIAPECAD E ARMAZENANDO EM UM DICIONÁRIO
        try:
            time.sleep(1)
            nav.switch_to.default_content()
            iframe = nav.find_element(By.NAME,'WA0') 
            nav.switch_to.frame(iframe)
            # capturando texto dos elementos (siapecad, nome_servidor)
            div_mscad = nav.find_element(By.ID, 'S_88').text
            div_nome_serv = nav.find_element(By.ID, 'S_127').text 
            dicionario_siapeCAD[div_nome_serv] = div_mscad
            nav.find_element(By.ID, 'B_2989').click()
        except:
            try:
                time.sleep(1)
                nav.switch_to.default_content()
                iframe = nav.find_element(By.NAME,'WA1') 
                nav.switch_to.frame(iframe)
                # capturando texto dos elementos (siapecad, nome_servidor)
                div_mscad = nav.find_element(By.ID, 'S_88').text
                div_nome_serv = nav.find_element(By.ID, 'S_127').text 
                dicionario_siapeCAD[div_nome_serv] = div_mscad
                nav.find_element(By.ID, 'B_2989').click()      
            except:
                time.sleep(1)
                nav.switch_to.default_content()
                iframe = nav.find_element(By.NAME,'WA2') 
                nav.switch_to.frame(iframe)
                # capturando texto dos elementos (siapecad, nome_servidor)
                div_mscad = nav.find_element(By.ID, 'S_88').text
                div_nome_serv = nav.find_element(By.ID, 'S_127').text 
                dicionario_siapeCAD[div_nome_serv] = div_mscad
                nav.find_element(By.ID, 'B_2989').click()
    except:
        try:    
            
            nav.switch_to.default_content()
            iframe = nav.find_element(By.NAME,'WA1') 
            nav.switch_to.frame(iframe)
            time.sleep(1)
            nav.find_element(By.ID, 'F_131').click()
            # # escrever a matrícula
            nav.find_element(By.XPATH, '//*[@id="F_131"]').send_keys(str(matricula).strip())
            # # precionar enter depois de colocar a matrícula
            ActionChains(nav).send_keys(Keys.ENTER).perform()
            time.sleep(1)
            # # pausa para o sistema 'pensar'
            time.sleep(1)
            # # PRESCIONANDO 'ENTER' PARA AVANÇAR A PÁGINA
            ActionChains(nav).send_keys(Keys.SPACE).perform()
            time.sleep(1)
            ActionChains(nav).send_keys(Keys.ENTER).perform()
            time.sleep(3)

            # COPIANDO NOME E MATRÍCULA SIAPECAD E ARMAZENANDO EM UM DICIONÁRIO
            try:
                time.sleep(1)
                nav.switch_to.default_content()
                iframe = nav.find_element(By.NAME,'WA0') 
                nav.switch_to.frame(iframe)
                # capturando texto dos elementos (siapecad, nome_servidor)
                div_mscad = nav.find_element(By.ID, 'S_88').text
                div_nome_serv = nav.find_element(By.ID, 'S_127').text 
                dicionario_siapeCAD[div_nome_serv] = div_mscad
                nav.find_element(By.ID, 'B_2989').click()
            except:
                try:
                    time.sleep(1)
                    nav.switch_to.default_content()
                    iframe = nav.find_element(By.NAME,'WA1') 
                    nav.switch_to.frame(iframe)
                    # capturando texto dos elementos (siapecad, nome_servidor)
                    div_mscad = nav.find_element(By.ID, 'S_88').text
                    div_nome_serv = nav.find_element(By.ID, 'S_127').text 
                    dicionario_siapeCAD[div_nome_serv] = div_mscad
                    nav.find_element(By.ID, 'B_2989').click()
                except:
                    time.sleep(1)
                    nav.switch_to.default_content()
                    iframe = nav.find_element(By.NAME,'WA2') 
                    nav.switch_to.frame(iframe)
                    # capturando texto dos elementos (siapecad, nome_servidor)
                    div_mscad = nav.find_element(By.ID, 'S_88').text
                    div_nome_serv = nav.find_element(By.ID, 'S_127').text 
                    dicionario_siapeCAD[div_nome_serv] = div_mscad
                    nav.find_element(By.ID, 'B_2989').click()
            
        except:
            time.sleep(1)
            nav.switch_to.default_content()
            iframe = nav.find_element(By.NAME,'WA2') 
            nav.switch_to.frame(iframe)
            time.sleep(1)
            nav.find_element(By.ID, 'F_131').click()
            # # escrever a matrícula
            nav.find_element(By.XPATH, '//*[@id="F_131"]').send_keys(str(matricula).strip())
            # # precionar enter depois de colocar a matrícula
            ActionChains(nav).send_keys(Keys.ENTER).perform()
            time.sleep(1)
            # # pausa para o sistema 'pensar'
            time.sleep(1)
            # # PRESCIONANDO 'ENTER' PARA AVANÇAR A PÁGINA
            ActionChains(nav).send_keys(Keys.SPACE).perform()
            time.sleep(1)
            ActionChains(nav).send_keys(Keys.ENTER).perform()
            time.sleep(3)
            
            # COPIANDO NOME E MATRÍCULA SIAPECAD E ARMAZENANDO EM UM DICIONÁRIO
            try:
                time.sleep(1)
                nav.switch_to.default_content()
                iframe = nav.find_element(By.NAME,'WA0') 
                nav.switch_to.frame(iframe)
                # capturando texto dos elementos (siapecad, nome_servidor)
                div_mscad = nav.find_element(By.ID, 'S_88').text
                div_nome_serv = nav.find_element(By.ID, 'S_127').text 
                dicionario_siapeCAD[div_nome_serv] = div_mscad
                nav.find_element(By.ID, 'B_2989').click()
            except:
                try:
                    time.sleep(1)
                    nav.switch_to.default_content()
                    iframe = nav.find_element(By.NAME,'WA1') 
                    nav.switch_to.frame(iframe)
                    # capturando texto dos elementos (siapecad, nome_servidor)
                    div_mscad = nav.find_element(By.ID, 'S_88').text
                    div_nome_serv = nav.find_element(By.ID, 'S_127').text 
                    dicionario_siapeCAD[div_nome_serv] = div_mscad
                    nav.find_element(By.ID, 'B_2989').click()
                except:
                    time.sleep(1)
                    nav.switch_to.default_content()
                    iframe = nav.find_element(By.NAME,'WA2') 
                    nav.switch_to.frame(iframe)
                    # capturando texto dos elementos (siapecad, nome_servidor)
                    div_mscad = nav.find_element(By.ID, 'S_88').text
                    div_nome_serv = nav.find_element(By.ID, 'S_127').text 
                    dicionario_siapeCAD[div_nome_serv] = div_mscad
                    nav.find_element(By.ID, 'B_2989').click()

    # A cada 20 registros grava no arquivo                
    if (len(dicionario_siapeCAD) > 0) and (len(dicionario_siapeCAD) % 20 == 0):
        with open('siapecadServidores.csv', 'a', encoding='utf-8') as scad:
            # scad.write(f'SIAPECAD,NOME\n')
            for chave in dicionario_siapeCAD:
                scad.write(f'{dicionario_siapeCAD[chave]},{chave}\n')
        dicionario_siapeCAD.clear()                
                    
           
print('Servidores registrados: ', len(dicionario_siapeCAD))
```
